```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Exploratory Data Analysis

```python

#Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/IRIS.csv")

#Show the first few rows of the dataset
data.head(10)
```

```python

# Scatterplot 1: Sepal Length vs Petal Length
sns.scatterplot(data=data, x="sepal_length", y="petal_length", hue="species", palette="magma")
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()

# Scatterplot 2: Sepal Width vs Petal Width
sns.scatterplot(data=data, x="sepal_width", y="petal_width", hue="species", palette="viridis")
plt.title("Sepal Width vs. Petal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species")
plt.show()
```

**What can you say about the relationship given the figure?**
The figure above reveals a distinct positive linear relationship between the sepal and petal dimensions for the species Versicolor and Virginica. However, for Setosa, there is no clear trend between these dimensions. This suggests that while sepal and petal dimensions are correlated, the nature of this relationship varies depending on the species.

```python
#Skapa boxplots fr varje numerisk variabel, grupperat efter species
numerical_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

#Iterate through each numerical column and create a boxplot.
for column in numerical_columns:
    sns.boxplot(data=data, x="species", y=column, hue="species", palette="viridis", dodge=False)
    plt.figure(figsize=(8, 5))
```

**How are the sepal and petal dimensions distributed?**

Based on the figures, we can observe that the sepal width is fairly consistent across all three species, showing little variation. However, for sepal length, petal length, and petal width, Iris virginica consistently has the largest values, followed by Iris versicolor, with Iris setosa having the smallest. Outliers are present in the data but are not particularly extreme.

Additionally, it is notable that Iris setosa shows more variation in sepal width compared to the other two species, while sepal length seems to exhibit similar variability across all three species. These observations highlight distinct differences in the size and variability of floral dimensions among the species.

```python
# Create a pairs plot
sns.pairplot(data, hue='species', diag_kind='kde', palette='viridis')

# Show the plot
plt.show()
```

**Briefly, mention how the different variables are related to each other.**

There seems to be a noticeable linear relationship between petal length and petal width, suggesting that as one increases, the other tends to do so as well. Similarly, sepal length and petal width also show a somewhat linear association, indicating a potential connection between these dimensions across the species.

# 2. Birdwatching

```python
#Load the data
data = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/artportalen.csv")

#Show the first few rows of the data
data.head(10)
```

```python

# Convert 'Antal' to string type to safely apply string operations
data['Antal'] = data['Antal'].astype(str)

# Extract only numeric values from 'Antal' and convert to numeric
data['Antal'] = pd.to_numeric(data['Antal'].str.extract(r'(\d+)', expand=False), errors='coerce')

# Remove rows where 'Antal' is NaN
data = data.dropna(subset=['Antal'])

# Group data by 'Artnamn' and calculate the total counts
species_counts = (
    data.groupby('Artnamn')
    .agg({
        'Antal': 'sum'  # Sum of counts for each species
    })
    .reset_index()
)

# Sort by the total count in descending order
species_counts = species_counts.sort_values('Antal', ascending=False)

# Display the most prevalent species
print("The most prevalent species are:")
species_counts.head(10)


```

```python

# Convert 'Startdatum' to datetime to extract month and year
data['Startdatum'] = pd.to_datetime(data['Startdatum'], errors='coerce')

# Ensure 'Antal' is numeric (convert any non-numeric values to NaN)
data['Antal'] = pd.to_numeric(data['Antal'], errors='coerce')

# Extract month and year from 'Startdatum'
data['YearMonth'] = data['Startdatum'].dt.to_period('M')

# Get the top 3 most prevalent species from species_counts
top_3_species = species_counts.head(3)['Artnamn']

# Filter data for these top 3 species
top_3_data = data[data['Artnamn'].isin(top_3_species)]

# Group by 'YearMonth' and 'Artnamn' and sum 'Antal'
monthly_distribution = top_3_data.groupby(['YearMonth', 'Artnamn'])['Antal'].sum().unstack().fillna(0)


# Visualize the monthly distribution
monthly_distribution.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'pink', 'brown'])
plt.title('Monthly Distribution of the Top 3 Most Prevalent Species')
plt.xlabel('Month')
plt.ylabel('Number of Sightings')
plt.xticks(rotation=45)
plt.legend(title='Species')
plt.tight_layout()
plt.show()

```

```python
#Sort by 'Antal' in ascending order to find the rarest species.
rarest_species = species_counts.sort_values(by='Antal', ascending=True)

#Show the rarest species
print("The rarest species are:")
rarest_species.head(20) 
```

**1. How does the number of bird sightings vary across different regions (Ln, Kommun, Frsamling)?**

This question is significant as it helps identify regions with high bird diversity, critical habitats, or gaps in data collection. It can guide conservation efforts, highlight the impact of urbanization, and reveal patterns in citizen science participation. Understanding regional differences supports better environmental planning and community engagement.


```python
#Extract only numeric values from 'Antal'.
data['Antal'] = data['Antal'].astype(str).str.extract('(\d+)', expand=False)

#Convert to numeric type
data['Antal'] = pd.to_numeric(data['Antal'], errors='coerce')

data = data.dropna(subset=['Antal'])

region_counts = data.groupby(['Ln', 'Kommun', 'Frsamling'])['Antal'].sum().reset_index()
region_counts = region_counts.sort_values(by='Antal', ascending=False)
region_counts.head()


```

**2. Is there a relationship between the age/stage (lder/stadium) of birds and the number of sightings (Antal)?**

This question explores whether certain stages in a birds life cycle (e.g., juvenile vs adult) lead to more sightings. It might help understand bird behavior, migration patterns, or nesting habits, which can be crucial for conservation efforts.


```python

# Step 1: Replace 'k', 'M', 'B' with numeric values (handling case insensitivity)
data['Antal'] = data['Antal'].replace({'k': 'e3', 'K': 'e3', 'M': 'e6', 'B': 'e9'}, regex=True)

# Step 2: Convert the 'Antal' column to numeric
data['Antal'] = pd.to_numeric(data['Antal'], errors='coerce')

# Step 3: Handle missing or NaN values (optional: fill with 0 or another method)
data['Antal'] = data['Antal'].fillna(0)

# Step 4: Group by 'lder/stadium' and sum 'Antal'
age_stage_counts = data.groupby('lder/stadium')['Antal'].sum().reset_index()

# Step 5: Sort by the number of sightings
age_stage_counts = age_stage_counts.sort_values(by='Antal', ascending=False)

# Step 7: Optional: Visualize the distribution
age_stage_counts.plot(kind='bar', x='lder/stadium', y='Antal', figsize=(10, 6), title="Bird Sightings by Age/Stage")
```

**3. Is there sertain times during the day where the number sightings are higher?**

The time of day can have a significant impact on animal behavior, including birds. Certain species may be more active during specific times of the day, such as dawn or dusk. Understanding if there is a relationship between sightings and time can help with planning more effective birdwatching excursions or provide insight into animal behavior patterns throughout the day.


```python

# First, convert 'Starttid' to datetime if it isn't already
data['Starttid'] = pd.to_datetime(data['Starttid'], errors='coerce')

# Extract the hour of the day from 'Starttid'
data['Hour'] = data['Starttid'].dt.hour

# Group by the hour of the day and sum the 'Antal' (Number of sightings)
time_of_day_counts = data.groupby('Hour')['Antal'].sum().reset_index()

# Sort by hour
time_of_day_counts = time_of_day_counts.sort_values(by='Hour')


```

```python
# Visualize the results using a line plot
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.lineplot(data=time_of_day_counts, x='Hour', y='Antal')
plt.title("Species Sightings by Time of Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Sightings")
plt.show()
```

By analyzing the distribution of sightings across different times of day, we can observe peak activity periods. We see that around 8 and 13 we have the most sightings,. Conversely, the data show lower activity during midday so then the birds are less active. This is interesting because the information be used to guide birdwatching effort and optimize the timing of future observations. 


# 3. Predicting Strokes

```python

# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/stroke-data.csv")

# Print the columns of the dataset
print(data.columns)

# Quick overview of the dataset
print(data.info())
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Stroke occurrence
sns.countplot(x='stroke', data=data)
plt.title("Stroke Occurrence")
plt.xlabel("Stroke (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# Stroke occurrence by age and hypertension
sns.countplot(data=data, x='age', hue='stroke')
plt.title("Age status distribution by Stroke Occurrence")
plt.show()

# Smoking status and stroke
sns.countplot(data=data, x='smoking_status', hue='stroke')
plt.title("Smoking Status Distribution by Stroke Occurrence")
plt.show()

# Stroke and gender
sns.countplot(data=data, x='gender', hue='stroke')
plt.title("Gender status distribution by Stroke Occurrence")
plt.show()

# BMI and stroke
sns.countplot(data=data, x='bmi', hue='stroke')
plt.title("BMI status distribution by Stroke Occurrence")
plt.show()

# Residence type and stroke
sns.countplot(data=data, x='Residence_type', hue='stroke')
plt.title("Residence type status distribution by Stroke Occurrence")
plt.show()

# Marital status and stroke
sns.countplot(data=data, x='ever_married', hue='stroke')
plt.title("Marriage status distribution by Stroke Occurrence")
plt.show()

# -------------------------------------------
# Correlation Analysis
# -------------------------------------------

# First, let's filter out non-numeric columns for correlation analysis
correlation_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Correlation Matrix of Numerical Features")
plt.show()




```

Upon analyzing the dataset, it is evident that the number of individuals who have not experienced a stroke vastly outweighs those who have. This is apparent in the imbalance of the stroke column, where the majority of records are labeled as 0 (no stroke). Such an imbalance might lead to a skewed analysis if we only consider the data as-is, especially in classification tasks, where the model may become biased towards predicting "no stroke" due to the overwhelming number of negative cases. We see that we dont have any correlation between any variable and stroke except for age who has a weak correlation.

We see that from this smokeing does not have a significant inpact on the chance of having a stroke

We see a weak correlation between age and stroke, and from the plot we see that in the increase of age we have a increase of amouts of strokes.
